```python
# Lib required  

import pandas as pd
import pandas as pd
import ast
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
#Loading the individual datasets 

data = pd.read_csv('songs.csv')
data_artist = pd.read_csv('artist-genres.csv')
data_genre = pd.read_csv('genres.csv')
data_year = pd.read_csv('year.csv')
data_artists = pd.read_csv('artist.csv')
```

```python
data.shape
```

```python
# One-hot encoding 'key' attribute

data_key = data[['key']]
data_key_onehot = OneHotEncoder(dtype=np.int8).fit_transform(data_key).toarray()
```

```python
# Dropping original 'key' column 

data.drop(['key'], axis=1, inplace=True)
key_indices = [f'key_{i}' for i in range(12)]
print(key_indices)
```

```python
data = data.join(pd.DataFrame(data_key_onehot, columns=key_indices))
```

```python
data.head()
```

```python
# One-hot encoding 'mode' attribute

data_mode = data[['mode']]
data_mode_onehot = OneHotEncoder(dtype=np.int8).fit_transform(data_mode).toarray()
```

```python
# Dropping original 'mode' column and joining encoded modes

data.drop(['mode'], axis=1, inplace=True)
data = data.join(pd.DataFrame(data_mode_onehot, columns=['minor', 'major']))
```

```python
df = data.copy()
```

```python
# One-hot encoding artist keys

data_artist_key = data_artist[['key']]
data_artist_key_onehot = OneHotEncoder(dtype=np.int8).fit_transform(data_artist_key).toarray()
data_artist.drop(['key'], axis=1, inplace=True)
```

```python
len(data_artist_key_onehot)
```

```python
# Joining artist key encodings and encoding artist modes

data_artist = data_artist.join(pd.DataFrame(data_artist_key_onehot, columns=key_indices))
data_artist_mode = data_artist[['mode']]
data_artist_mode_onehot = OneHotEncoder(dtype=np.int8).fit_transform(data_artist_mode).toarray()
```

```python
data_artist = data_artist.join(pd.DataFrame(data_artist_mode_onehot, columns=['minor', 'major']))
```

```python
# Displaying first few rows of artist data to verify encoding

data_artist.head()
```

```python
# One-hot encoding genre keys and joining

data_genre_key = data_genre[['key']]
data_genre_key_onehot = OneHotEncoder(dtype=np.int8).fit_transform(data_genre_key).toarray()
data_genre.drop(['key'], axis=1, inplace=True)
data_genre = data_genre.join(pd.DataFrame(data_genre_key_onehot, columns=key_indices))
```

```python
# Joining artist and genre data

data_artist = data_artist.join(data_genre.set_index('genres'), on='genres', rsuffix='_gr')
```

```python
# Exploding 'artists' list to rows and setting index for joining

data['artists'] = data['artists'].apply(ast.literal_eval)
data = data.explode('artists')
data_artist = data_artist.set_index('artists')
data.tail(25)
```

```python
# Joining exploded artist data with song data

data = data.join(data_artist, on='artists', rsuffix='_ar')
```
